```python
import pandas as pd
import numpy as np
import os
import sys
from collections import OrderedDict
```

```python
def openDataFile(name):
    entries = os.listdir('data/')
    i = 1
    for v in entries:
        print(str(i) + '.'+v)
        i = i + 1
    print('Seleccione el numero del archivo: ', name)
    print('En caso de no tener el archivo en data puede ingresar el directorio')
    aux = input()
    if len(aux) <= 5:
        i = 1
        for v in entries:
            if aux == str(i):
                return pd.read_excel('data/' + v)
            i = i + 1
    else: 
        return pd.read_excel(v)
```

```python
def setCanal(length, db):
    global setActual
    global lineaActual
    setActual = 'Canal'
    lineaActual = 1
    canal = []
    canalInt = []
    canalMa = []
    fabricante = []
    for x in range(length):        
        fabricante.append('HEINEKEN')
        canal.append(db['Cliente Cadena'][x])
        canalInt.append('Off Modern')
        canalMa.append('OTHER -CS + PROX')
        lineaActual = lineaActual + 1
    return canal, canalInt, canalMa, fabricante
```

```python
def setGZ(gz, gzCat):
    global setActual
    global lineaActual
    setActual = 'GZ'
    lineaActual = 1
    region = []
    gz1 = []
    fabricante = []
    for value in gz:
        loc = gzCat.loc[gzCat.GZ == value]
        lineaActual = lineaActual + 1
        region.append(loc['REGION'].iloc[0])
        gz1.append(loc['GZ.1'].iloc[0])
    return region, gz1
```

```python
def setUE(multipacks, emCat):
    global setActual
    global lineaActual
    setActual = 'Pack'
    lineaActual = 1
    unidadesDeEmpaque = []
    for value in multipacks:
        loc = emCat.loc[emCat.Multipacks == value]
        lineaActual = lineaActual + 1
        unidadesDeEmpaque.append(loc['Val'].iloc[0])
    return unidadesDeEmpaque
```

```python
def setReturnability(ez, returnabilityCat):
    global setActual
    global lineaActual
    setActual = 'Returnability'
    lineaActual = 1
    returnability = []
    primaryPackType = []
    primaryPackGroup = []
    hMexPack = []
    packSize = []
    packtypeHOM = []
    primaryPacktypeGroup = []
    groupPacksize = []
    packSizeHOM = []
    packtypeOk2 = []
    categorias = []
    packSize1Home = []
    pConcatenado1 = []
    pConcatenado2 = []
    presentacion = []
    
    for value in ez:    
        loc = returnabilityCat.loc[returnabilityCat['PresentaciÃ³n CCM'] == value]
        returnability.append(loc['Returnability'].iloc[0])
        if loc['Returnability'].iloc[0] == 'Non Returnable':
            presentacion.append('NO RETORNABLE')
        else: 
            presentacion.append('RETORNABLE')
        primaryPackType.append(loc['Primary Pack Type'].iloc[0])
        primaryPackGroup.append(loc['Primary Pack Group'].iloc[0])
        hMexPack.append(loc['HMex Pack Group'].iloc[0])
        packSize.append(loc['Pack Size'].iloc[0])
        packtypeHOM.append(loc['Packtype_HOM'].iloc[0])
        primaryPacktypeGroup.append(loc['Primary_Packtype_Group'].iloc[0])
        groupPacksize.append(loc['Group_Packsize'].iloc[0])
        packSizeHOM.append(loc['Packsize_HOM'].iloc[0])
        packtypeOk2.append(loc['PACKTYPE_OK_2'].iloc[0])
        categorias.append(loc['Categorias'].iloc[0])
        packSize1Home.append(loc['Packsize1_HOME'].iloc[0])
        pConcatenado1.append(loc['Packsize_concatenado1'].iloc[0])
        pConcatenado2.append(loc['Packsize_concatenado2'].iloc[0])
        lineaActual = lineaActual + 1 
       
    return returnability, primaryPackType, primaryPackGroup, hMexPack, packSize, packtypeHOM, primaryPacktypeGroup, groupPacksize, packSizeHOM, packtypeOk2, categorias, packSize1Home, pConcatenado1, pConcatenado2, presentacion
```

```python
def errorMessage(setActual, lineaActual): 
    print('ERROR')
    print('Se ha detectado un valor inexistente en el Catalogo', setActual)
    print('Revisar en el documento de datos en la linea: ', lineaActual)
    print('Favor de agregar el valor faltante a el catalago y volver a intentar')
    print('Ingrese ENTER para continuar')
    print('Ingrese cualquier otra tecla para salir')
    aux = input()
    if not aux:
        return True
    else: 
        sys.exit()
```

```python
def fixDateValues(monthyValues):
    year = []
    column = monthyValues.columns.tolist()
    for month in column:
        year.append(month[-2:])
    res = list(OrderedDict.fromkeys(year))
    return res
    
```

```python
def setMarca(le, marcaCat):
    global setActual
    global lineaActual
    setActual = 'Marca'
    lineaActual = 1
    marca_hom = []
    ext_marca_hom = []
    diamante_precio_est = []
    tipo_liq_est = []
    liq = []
    marca_hom2 = []
    ext_marca_hom2 = []
    diamante_precio_hom = []
    manu = []
    priceTierCMI = []
    categoryCMI = []
    alcoholSegmentsCMI = []
    liquidSegmentsCMI = []
    brandCondition = []
    topLineSOg = []
    secSG = []
    
    for value in le:    
        loc = marcaCat.loc[marcaCat['Line Extension'] == value]
        lineaActual = lineaActual + 1
        marca_hom.append(loc['MARCA_HOM'].iloc[0])
        ext_marca_hom.append(loc['EXT_MARCA_HOM'].iloc[0])
        diamante_precio_est.append(loc['DIAMANTE_PRECIO_EST'].iloc[0])
        tipo_liq_est.append(loc['TIPO_LIQ_EST'].iloc[0])
        liq.append(loc['LIQ'].iloc[0])       
        marca_hom2.append(loc['MARCA_HOM.1'].iloc[0])
        ext_marca_hom2.append(loc['EXT_MARCA_HOM.1'].iloc[0])
        diamante_precio_hom.append(loc['DIAMANTE_PRECIO_HOM'].iloc[0])
        manu.append(loc['Manufacturer_CMI'].iloc[0])
        priceTierCMI.append(loc['Price_Tier_CMI'].iloc[0])
        categoryCMI.append(loc['Category_CMI'].iloc[0])  
        alcoholSegmentsCMI.append(loc['Alcohol_Segments_CMI'].iloc[0])
        liquidSegmentsCMI.append(loc['Liquid_Segment_CMI'].iloc[0])
        brandCondition.append(loc['BRAND_CONDITION'].iloc[0])
        topLineSOg.append(loc['TOP LINE SOG (5 SOGS)'].iloc[0])
        secSG.append(loc['Secondary_Source_of_Growth'].iloc[0])
        lineaActual = lineaActual + 1
    return marca_hom, ext_marca_hom, diamante_precio_est, tipo_liq_est, liq, marca_hom2, ext_marca_hom2, diamante_precio_hom, manu, priceTierCMI,categoryCMI, alcoholSegmentsCMI, liquidSegmentsCMI, brandCondition, topLineSOg, secSG
```

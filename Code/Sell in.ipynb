```python
def setCanal(canal, canalCat):
    global setActual
    global lineaActual
    global valor
    setActual = 'Canal'
    lineaActual = 1
    canalInt = []
    canalMa = []
    fabricante = []
    for value in canal:
        valor = value
        loc = canalCat.loc[canalCat.Canal == value]
        fabricante.append('HEINEKEN')
        canalInt.append(loc['CANAL INTEGRADO'].iloc[0])
        canalMa.append(loc['CANAL_MA'].iloc[0])
        lineaActual = lineaActual + 1
    return canalInt, canalMa, fabricante
```

```python
def setGZ(gz, gzCat):
    global setActual
    global lineaActual
    global valor
    setActual = 'GZ'
    lineaActual = 1
    region = []
    gz1 = []
    fabricante = []
    for value in gz:
        valor = value
        lineaActual = lineaActual + 1
        loc = gzCat.loc[gzCat.GZ == value]
        region.append(loc['REGION'].iloc[0])
        gz1.append(loc['GZ.1'].iloc[0])
    return region, gz1
```

```python
def setPz(dz, pzCat):
    global setActual
    global lineaActual
    global valor
    setActual = 'PZ'
    lineaActual = 1
    pz = []
    for value in dz:
        valor = value
        lineaActual = lineaActual + 1
        loc = pzCat.loc[pzCat.Llave == value]
        pz.append(loc.PZ.iloc[0])
    return pz
```

```python
def setMarca(le, marcaCat):
    global setActual
    global lineaActual
    global valor
    setActual = 'Marca'
    lineaActual = 1
    marca_hom = []
    ext_marca_hom = []
    diamante_precio_est = []
    tipo_liq_est = []
    liq = []
    marca_hom2 = []
    ext_marca_hom2 = []
    diamante_precio_hom = []
    manu = []
    priceTierCMI = []
    categoryCMI = []
    alcoholSegmentsCMI = []
    liquidSegmentsCMI = []
    brandCondition = []
    topLineSOg = []
    mainstreamSegmentation = []
    sesionable = []
    for value in le:
        valor = value
        lineaActual = lineaActual + 1
        loc = marcaCat.loc[marcaCat['Line Extension'] == value]
        marca_hom.append(loc['MARCA_HOM'].iloc[0])
        ext_marca_hom.append(loc['EXT_MARCA_HOM'].iloc[0])
        diamante_precio_est.append(loc['DIAMANTE_PRECIO_EST'].iloc[0])
        tipo_liq_est.append(loc['TIPO_LIQ_EST'].iloc[0])
        liq.append(loc['LIQ'].iloc[0])
        marca_hom2.append(loc['MARCA_HOM.1'].iloc[0])
        ext_marca_hom2.append(loc['EXT_MARCA_HOM.1'].iloc[0])
        diamante_precio_hom.append(loc['DIAMANTE_PRECIO_HOM'].iloc[0])
        manu.append(loc['Manufacturer_CMI'].iloc[0])
        priceTierCMI.append(loc['Price_Tier_CMI'].iloc[0])
        categoryCMI.append(loc['Category_CMI'].iloc[0])
        alcoholSegmentsCMI.append(loc['Alcohol_Segments_CMI'].iloc[0])
        liquidSegmentsCMI.append(loc['Liquid_Segment_CMI'].iloc[0])
        brandCondition.append(loc['BRAND_CONDITION'].iloc[0])
        topLineSOg.append(loc['TOP LINE SOG (5 SOGS)'].iloc[0])
        mainstreamSegmentation.append(loc['MAINSTREAM SEGMENTATION'].iloc[0])
        sesionable.append(loc['SESSIONABLE LIGHT/UTRA SOG'].iloc[0])
    return marca_hom, ext_marca_hom, diamante_precio_est, tipo_liq_est, liq, marca_hom2, ext_marca_hom2, diamante_precio_hom, manu, priceTierCMI,categoryCMI, alcoholSegmentsCMI, liquidSegmentsCMI, brandCondition, topLineSOg, mainstreamSegmentation, sesionable
```

```python
def setReturnability(ez, returnabilityCat):
    global setActual
    global lineaActual
    global valor
    setActual = 'Returnability'
    lineaActual = 1
    returnability = []
    llave = []
    primaryPackType = []
    primaryPackGroup = []
    hMexPack = []
    packSize = []
    packtypeHOM = []
    primaryPacktypeGroup = []
    groupPacksize = []
    packSizeHOM = []
    packtypeOk2 = []
    categorias = []
    packSize1Home = []
    pConcatenado1 = []
    pConcatenado2 = []
    
    for value in ez:   
        valor = value
        loc = returnabilityCat.loc[returnabilityCat.llave == value]
        llave.append(value)
        returnability.append(loc['Returnability'].iloc[0])
        primaryPackType.append(loc['Primary Pack Type'].iloc[0])
        primaryPackGroup.append(loc['Primary Pack Group'].iloc[0])
        hMexPack.append(loc['HMex Pack Group'].iloc[0])
        packSize.append(loc['Pack Size'].iloc[0])
        packtypeHOM.append(loc['Packtype_HOM'].iloc[0])
        primaryPacktypeGroup.append(loc['Primary_Packtype_Group'].iloc[0])
        groupPacksize.append(loc['Group_Packsize'].iloc[0])
        packSizeHOM.append(loc['Packsize_HOM'].iloc[0])
        packtypeOk2.append(loc['PACKTYPE_OK_2'].iloc[0])
        categorias.append(loc['Categorias'].iloc[0])
        packSize1Home.append(loc['Packsize1_HOME'].iloc[0])
        pConcatenado1.append(loc['Packsize_concatenado1'].iloc[0])
        pConcatenado2.append(loc['Packsize_concatenado2'].iloc[0])
        lineaActual = lineaActual + 1 
       
    return llave, returnability, primaryPackType, primaryPackGroup, hMexPack, packSize, packtypeHOM, primaryPacktypeGroup, groupPacksize, packSizeHOM, packtypeOk2, categorias, packSize1Home, pConcatenado1, pConcatenado2
```

```python
def errorMessage(setActual, lineaActual, valor): 
    print('ERROR')
    print('Se ha detectado un valor inexistente en el Catalogo', setActual)
    print('Revisar en el documento de datos en la linea: ', lineaActual)
    print('El valor no registrado es: ', valor)
    print('Favor de agregar el valor faltante a el catalago y volver a intentar')
    print('Ingrese ENTER para continuar')
    print('Ingrese cualquier otra tecla para salir')
    aux = input()
    if not aux:
        return True
    else: 
        sys.exit()
```

```python
def printTable(db, fabricante, canalInt, canalMA, region, gz1, pz, marca_hom, ext_marca_hom, diamante_precio_est, tipo_liq_est, liq, llave, returnability, primaryPackType, primaryPackGroup, hMexPack, packSize, marca_hom2, ext_marca_hom2, diamante_precio_hom, groupPacksize, packtypeOk2, categorias, packSize1Home, pConcatenado1, pConcatenado2, manu, priceTierCMI, categoryCMI, alcoholSegmentsCMI, liquidSegmentsCMI, brandCondition, topLineSOg, mainstreamSegmentation, sesionable):
    data = [db.Year, db.Month, db.Quarter, fabricante, db.Channel, canalInt, canalMA, db.DR, db['GZ Split TVN-VDM'], region, gz1, pz, db.Brand, db['Line Extension'], marca_hom, ext_marca_hom, diamante_precio_est, tipo_liq_est, liq, db['Price Segment'], llave, returnability, primaryPackType, primaryPackGroup, hMexPack, packSize, marca_hom2, ext_marca_hom2, diamante_precio_hom, llave, returnability, primaryPackType, primaryPackGroup, groupPacksize, packSize, packtypeOk2, categorias, packSize1Home, pConcatenado1, pConcatenado2, manu, priceTierCMI, categoryCMI, alcoholSegmentsCMI, liquidSegmentsCMI, brandCondition, topLineSOg, mainstreamSegmentation, sesionable, db.Multipack, db.SKU, db.Value]
    names = ['AÑO', 'MES', 'QUARTER','FABRICANTE','CANAL', 'CANAL INTEGRADO', 'CANAL_MA', 'DR','GZ', 'REGION', 'GZ', 'PZ','BRAND', 'LINE EXTENSION', 'MARCA_HOM', 'EXT_MARCA_HOM', 'DIAMANTE_PRECIO_EST', 'TIPO_LIQUIDO_EST', 'LIQ', 'PRICE_SEGMENT', 'LLAVE','Returnability', 'Packtype_HOM', 'Primary_Packtype_Group', 'HMEX_PACK_GROUP', 'PACK_SIZE','MARCA_HOM2', 'EXT_MARCA_HOM2', 'DIAMANTE_PRECIO_HOM', 'LLAVE2','RETURNABILITY2', 'PRIMARY_PACK_TYPE2', 'PRIMARY_PACK_GROUP2', 'Group_Packsize', 'Packsize_HOM', 'PACKTYPE_OK_2', 'CATEGORIAS', 'Packsize1_HOM', 'Packsize_Concatenado1', 'Packsize_Concatenado2', 'Manufacturer_CMI', 'Price_Tier_CMI', 'Category_CMI', 'Alcohol_Segments_CMI', 'Liquid_Segments_CMI', 'BRAND_CONDITION', 'TOP LINE SOG (5 SOGS)', 'MAINSTREAM_SEGMENTATION', 'SESSIONABLE LIGHT/UTRA SOG', 'MULTIPACK', 'SKU', 'VOLUME']
    df = pd.DataFrame(data, names).T
    df.to_excel("Sell In.xlsx", index = False)
```

```python
def generateDatabase():
    
    try: 
        nombreDB, cat = SellInReadFile()
        db = pd.read_excel(nombreDB)
    ## Obtener todos los valores con la llave de canal
        global setActual
        global lineaActual
        global valor
        setActual = ''
        lineaActual = 999999
    except: 
        print('ERROR')
        print('Se ha generado un problema en la lectura del archivo')
        print('Los documentos esperados son ' + nombreDB + ' y ' + cat)
        print('Favor de revisar estos documentos y volver a intentar') 

    flag = True
    while flag:
        canalCat = pd.read_excel(cat, sheet_name = 'Canales')
        os.system('cls')
        try:    
            ## Obtener los valores con la llave de canal
            cain, cama, fa  = setCanal(db.Channel, canalCat)
            canalInt = pd.Series(cain)
            canalMA = pd.Series(cama)
            fabricante = pd.Series(fa)
            flag = False
        except:
            flag = errorMessage(setActual, lineaActual, valor)
            
    flag = True
    while flag:
        gzCat = pd.read_excel(cat, sheet_name = 'Gz')
        os.system('cls')
        try:
            ## Obtener todos los valores con la llave de GZ   
            re, g1 = setGZ(db['GZ Split TVN-VDM'], gzCat)
            region = pd.Series(re)
            gz1 = pd.Series(g1)
            flag = False
        except:
            flag = errorMessage(setActual, lineaActual, valor)
            
    flag = True
    while flag:
        pzCat = pd.read_excel(cat, sheet_name = 'Pz')
        os.system('cls')
        try:
            dz =  db.DR+db['GZ Split TVN-VDM']+db['OfV Description']
            p = setPz(dz, pzCat)
            pz = pd.Series(p)
            flag = False
        except:
            flag = errorMessage(setActual, lineaActual, valor)
    
    flag = True
    while flag:
        marcaCat = pd.read_excel(cat, sheet_name = 'Marca')
        os.system('cls')
        try:
            ## Obtener todos los valores con la llave de Line Extension (Marca)
            mh, emh, dpe, tlq, li, mh2, emh2, dph, man, ptCMI, cCMI, asCMI, lsCMI, bc, tp, ms, ses = setMarca(db['Line Extension'], marcaCat)
            marca_hom = pd.Series(mh)
            ext_marca_hom = pd.Series(emh)
            diamante_precio_est = pd.Series(dpe)
            tipo_liq_est = pd.Series(tlq)
            liq = pd.Series(li)
            marca_hom2 = pd.Series(mh2)
            ext_marca_hom2 = pd.Series(emh2)
            diamante_precio_hom = pd.Series(dph)
            manu = pd.Series(man)
            priceTierCMI = pd.Series(ptCMI)
            categoryCMI = pd.Series(cCMI)
            alcoholSegmentsCMI = pd.Series(asCMI)
            liquidSegmentsCMI = pd.Series(lsCMI)
            brandCondition = pd.Series(bc)
            topLineSOg = pd.Series(tp)
            mainstreamSegmentation = pd.Series(ms)
            sesionable = pd.Series(ses)
            flag = False
        except:
            flag = errorMessage(setActual, lineaActual, valor)
        
    flag = True
    while flag:
        returnabilityCat = pd.read_excel(cat, sheet_name = 'Returnability')
        os.system('cls')
        try:
            ## Obtener todos los valores con la llave de Returnability
            ez = db['Returnability'] + db['Primary Pack Type'] + db['Primary Pack Group'] + db['HMex Pack Group'] + db['Pack Size']
            lla, ret, prPT, prPG, hMex, ps, ptHOM, prPTG, gPS, psHOM, ptOK2, cat, ps1H, pc1, pc2 = setReturnability(ez, returnabilityCat)
            llave = pd.Series(lla)
            returnability  = pd.Series(ret)
            primaryPackType = pd.Series(prPT)
            primaryPackGroup = pd.Series(prPG)
            hMexPack = pd.Series(hMex)
            packSize = pd.Series(ps)
            packtypeHOM = pd.Series(ptHOM)
            primaryPacktypeGroup = pd.Series(prPTG)
            groupPacksize = pd.Series(gPS)
            packSizeHOM = pd.Series(gPS)
            packtypeOk2 = pd.Series(ptOK2)
            categorias = pd.Series(cat)
            packSize1Home = pd.Series(ps1H)
            pConcatenado1 = pd.Series(pc1)
            pConcatenado2 = pd.Series(pc2)
            flag = False
        except:
            flag = errorMessage(setActual, lineaActual, valor)
    
    try:
    ## Formatear todos los datos y crear un excel con ellos
        printTable(db, fabricante, canalInt, canalMA, region, gz1, pz, marca_hom, ext_marca_hom, diamante_precio_est, tipo_liq_est, liq, llave, returnability, primaryPackType, primaryPackGroup, hMexPack, packSize, marca_hom2, ext_marca_hom2, diamante_precio_hom, groupPacksize, packtypeOk2, categorias, packSize1Home, pConcatenado1, pConcatenado2, manu, priceTierCMI, categoryCMI, alcoholSegmentsCMI, liquidSegmentsCMI, brandCondition, topLineSOg, mainstreamSegmentation, sesionable)
        print('El documento se ha creado con exito!')
 
    except:
        print('ERROR')
        print('Se ha detectado un error en la creación del archivo, favor de volver a intentar')
        
```
